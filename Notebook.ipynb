### Missing values

```python
df.isnull().sum()
```

```python
## Dropping rows where missing value is more than 10% of the data set total rows
# Identify columns with more than 10% missing values
cols_to_drop = df.columns[df.isnull().sum() > len(df) * 0.1]

# Drop those columns
df_filtered = df.drop(columns=cols_to_drop)
```

```python
df_filtered.isnull().sum()
```

### Imputing null values

```python
# imputing all columns with the mean value
numerical_columns = df_filtered.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_columns:
    df_filtered[col] = df_filtered[col].fillna(df_filtered[col].mean())

```

```python
df_filtered.isnull().sum()
```

```python
sample_columns = df_filtered[numerical_columns].sample(n=9, axis=1)  

# Set up the figure with subplots (adjust rows and columns as needed)
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15,15))  # One row of subplots
axes = axes.flatten()  # Flatten to make indexing easier

# Loop through the sampled columns and plot the distribution
for i, column in enumerate(sample_columns.columns):
    # Plot histogram
    sns.histplot(sample_columns[column], kde=True, ax=axes[i], bins=20)  # Using kde=True to add a KDE curve
    axes[i].set_title(f'Distribution of {column}')
    axes[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels if necessary

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```
